```python
# Write a for loop that produces: ["transaction_id", "amount_", "date_recorded", "is_valid", "source_system"]. Use .strip(), .lower(), .replace() on each item.

cols = ["  Transaction ID ", "Amount ($)", "Date-Recorded", "IS_VALID", "  source System"]

columns = []
for col in cols:
    col = col.strip()
    col = col.lower()
    col = col.replace(" ", "_")
    col = col.replace("-", "_")
    col = col.replace("($)", "")
    columns.append(col)

print(columns)     # ['transaction_id', 'amount_', 'date_recorded', 'is_valid', 'source_system']
```

Q8. A log aggregator gives you a raw multi-field string where fields use inconsistent delimiters — some use |, some use :, some have extra spaces:
Using only string methods (no imports, no functions yet), write step-by-step code that produces this clean dictionary — building each value as a correctly typed Python variable:
{
  "date":     "2024-09-12",
  "level":    "warning",
  "event":    "high latency detected",
  "pipeline": "etl_orders",
  "latency":  4521
}

```python
raw = "  date : 2024-09-12  |  level : WARNING | event : high_latency_detected | pipeline : etl_orders | latency_ms : 4521  "

raw_headers = raw.strip()
print(raw_headers)            # date : 2024-09-12  |  level : WARNING | event : high_latency_detected | pipeline : etl_orders | latency_ms : 4521

headers = raw_headers.split('|')
print(headers)                # ['date : 2024-09-12  ', '  level : WARNING ', ' event : high_latency_detected ', ' pipeline : etl_orders ', ' latency_ms : 4521']

log_headers = []
for head in headers:
    head = head.strip()
    head = head.lower()
    head = head.replace(' : ', ':')
    head = head.replace('_', ' ')
    log_headers.append(head)
    
print(log_headers)      # ['date:2024-09-12', 'level:warning', 'event:high latency detected', 'pipeline:etl orders', 'latency ms:4521']
```

Q9. You are normalizing a batch of email addresses from a form that had no input validation. Each email has at least one of the following problems: extra whitespace, mixed case, a trailing period, or the domain is written as "[at]" instead of "@". Write step-by-step code to clean each one, then check it ends with ".com" or ".org" — print "valid" or "invalid" for each.

```python
emails = [
  "  Alice.Smith@Company.COM  ",
  "BOB.JONES[at]example.org.",
  "carol@data.com",
  "  DAVE[at]PIPELINE.COM"
]

new_emails = []
for email in emails:
  email = email.strip()
  email = email.lower()
  email = email.strip('.')
  email = email.replace('[at]', '@')
  if email.endswith((".com", ".org")):
    print(f"{email} valid")
  else:
    print(f"{email} invalid")
  new_emails.append(email)
  
print(new_emails)  # ['alice.smith@company.com', 'bob.jones@example.org', 'carol@data.com', 'dave@pipeline.com']
```

Q10. A fixed-width flat file (a legacy format still common in finance and banking) stores records where every field occupies an exact number of characters — no delimiter, just position. Each record looks like this, all concatenated into one long string:
"txn_001 payments 001500.00 20240801"
Field positions: txn_id = chars 0–6, channel = chars 8–15, amount = chars 17–25, date = chars 27–34.
Write code that slices each field out of the raw string, strips it, and produces a clean dict with amount as a float and date reformatted as "YYYY-MM-DD" (insert - at positions 4 and 6). Do this for all three records:
pythonrecords = [
  "txn_001 payments 001500.00 20240801",
  "txn_002 refunds  000299.50 20240802",
  "txn_003 transfers000099.99 20240803"
]

```python
records = [
    "txn_001 payments 001500.00 20240801",
    "txn_002 refunds  000299.50 20240802",
    "txn_003 transfers000099.99 20240803"
]

new_records = []
for record in records:
    txn_id = record[0:7].strip()
    channel = record[8:16].strip()
    amount = float(record[17:26].strip(''))
    date = record[27:35].strip()
    
    # Reformat date from YYYYMMDD to YYYY-MM-DD
    date_formatted = f"{date[0:4]}-{date[4:6]}-{date[6:8]}"
    
    # Build clean dict
    new_records.append({
        "txn_id": txn_id,
        "channel": channel,
        "amount": amount,
        "date": date_formatted
    })

print(new_records)
# [{'txn_id': 'txn_001', 'channel': 'payments', 'amount': 1500.0, 'date': '2024-08-01'}, {'txn_id': 'txn_002', 'channel': 'refunds', 'amount': 299.5, 'date': '2024-08-02'}, {'txn_id': 'txn_003', 'channel': 'transfer', 'amount': 99.99, 'date': '2024-08-03'}]
```
