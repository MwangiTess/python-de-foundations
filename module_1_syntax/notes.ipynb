### Syntax & Variables
- Type checking and conversion are critical tools. Use **type()** to inspect, and built-in constructors **( int(), str(), float(), bool() )** to cast.
- IN python, string is not just text - its an **object**. Objects have **methods**: they are functions that are attached to the object and operate on its data. Calling them using a dot notation.

#### String Methods

```python
# Strings are immutable, methods don't change the original string, they always return a new string.
# .strip() removes whitespaces

name = "  alice  "
name.strip()            # does not change name
print(name)             # still "  alice  " -unchanged
print(name.strip())     # now "alice"

name = name.strip()     # Must reassign to capture the result
print(name)             # now "alice"
```

##### 1. Remove Whitespaces
- They remove leading and/or trailing whitespaces(spaces, tabs\t, newlines\n). Does nothing to whtespace inside the string:  **.strip(), .lstrip(), .rstrip()**

```python
# The three variants
s = "   hello world   "

print(s.strip())    # "hello world" - both sides
print(s.lstrip())   # "hello world     " - left sides only
print(s.rstrip())   # "    hello world" - right sides only
```

```python
city_from_csv = "  London  "
city_from_db = "London"

print(city_from_csv == city_from_db)             # False - the spaces break it
print(city_from_csv.strip() == city_from_db)     # True
```

```python
# Stripping specific charachters (not just whitespace)
price = "###99.99###"
print(price.strip("#"))      # 99.99

ref = "--TXN-00421--"
print(ref.strip("-"))        # TXN-00421
```

##### 2. Normalize case
- Returns a new string with all the characters converted to lowercase or uppercase **.lower(), .upper()**

```python
city = "New York City"
print(city.lower())     # new york city
print(city.upper())     # NEW YORK CITY
```

```python
# Case differences cause silent data quality bugs. Two records that refer to the same thing will look different to python.
# In DE, it is advisable to normalize to the lower case before storing or comparing fields

status_a = "ACTIVE"
status_b = "Active"
status_c = "active"

print(status_a == status_b)     # False
print(status_a == status_c)     # False
print(status_b == status_c)     # False

print(status_a.lower() == status_b.lower())     # True
print(status_a.lower() == status_c.lower())     # True
print(status_b.lower() == status_c.lower())     # True
```

##### Title and Sentence Case
- When displaying names, city names, propernouns we use: **.title()**
- In sentences and descriptions we use: **.capitalize()**

```python
city = "new york city"
print(city.title())         # New York City - capitalizes every word
print(city.capitalize())    # New york city - Only the first word is capitalized

name = "mary o'brien"
print(name.title())         # Mary O'Brien - capitalizes after apostrophe too
print(name.capitalize())    # Mary o'brien - Only the first word is capitalized
```

##### 3. Substitute substrings
- **.replace()** What it does: Replaces every occurence of a target substring with a replacement string. Returns a new string.

```python
s = "first_name,last_name,date_of_birth"

print(s) # first_name,last_name,date_of_birth
print(s.replace("_", " "))  # first name,last name,date of birth
print(s.replace(",", " | "))  # first_name | last_name | date_of_birth


# Limit replacement
s = "aaa"
print(s.replace("a", "b"))      # bbb
print(s.replace("a", "b", 2))   # bba
```

```python
# Data engineering use cases:
# Clean currency symbols before casting to float

price = "$1,299.99"
clean = price.replace("$", "").replace(",", "")
print(clean)            # "1299.99"
print(type(clean))      # str
print(float(clean))     # 1299.99

# Normalize path separators
raw_path = "data\\raw\\sales\\2024"
path = raw_path.replace("\\", "/")

print(raw_path)     # data\raw\sales\2024
print(path)         # data/raw/sales/2024
```

##### 4. Tokenize a string into a list
- **.split**: It splits a string on a delimiter and returns a list of substrings.

```python
# Default split: splits on any whitespace
s = "hello world  foo  bar"
print(s.split())        # ['hello', 'world', 'foo', 'bar']
                        # Note: collapses multiple spaces automatically
                        
# Split on a specific delimeter
csv_line = "Alice,30,London,Engineer"
print(csv_line.split(","))      # ['Alice', '30', 'London', 'Engineer']

# Split on multi-character delimeter
log = "ERROR -- disk full -- node_7"
print(log.split(" -- "))        # ['ERROR', 'disk full', 'node_7']

# Limit the number of splits
s = "key=value=with=equals"
print(s.split("="))             # ['key', 'value', 'with', 'equals']
print(s.split("=", 1))          # ['key', 'value=with=equals']
```

```python
# Accessing parts after split
record = "txn_001,payments,1500.00,2024-08-01"
parts = record.split(",")

print(parts)        # ['txn_001', 'payments', '1500.00', '2024-08-01']
print(parts[0])     # txn_001
print(parts[1])     # payments
print(parts[2])     # 1500.00
print(parts[3])     # 2024-08-01
print(len(parts))   # The length of the list is 4
```

##### 5. The reverse of split
- **.join()**: Takes a list of strings and joins them into one string using the ceiling string as the separator. Exact inverse of **.split()**

```python
parts = ["Alice", "30", "London", "Engineer"]

print(", ".join(parts))     # Alice, 30, London, Engineer
print(" | ".join(parts))    # Alice | 30 | London | Engineer
print("\n".join(parts))     # Alice each item on its own line
```

```python
#  .join() requires all items in the list to be strings. If any item is an int or float, it will raise a TypeError
mixed = ["txn_001", 1500, "2024-08-01"]

print(",".join(mixed))  # TypeError: sequence item 1: expected str instance, int found

# To convert non-strings before joining
safe = [str(x) for x in mixed]
result = ",".join(safe)

print(result)           # txn_001,1500,2024-08-01
```

##### 6. Check beginnings and endings
- **.startswith() and .endswith()**: Return **True or False**. Used for routing, filtering, and validation

```python
filename = "sales_2024_Q3.csv"

print(filename.startswith("sales"))     # True
print(filename.startswith("report"))    # False
print(filename.endswith(".csv"))        # True
print(filename.endswith(".json"))       # False

# You can pass a tuple of options to check multiple cases at once.
filename = "sales_2024_Q3.csv"

print(filename.endswith((".csv", ".tsv", ".json")))     # True - matches .csv
```

```python
# Routing files to the right parser, filtering log lines by severity.
log_line = "ERROR: connection timeout on node_4"

print(log_line.startswith("ERROR"))         # True - route to error handler
print(log_line.startswith("INFO"))          # False
print(log_line.startswith("WARNING"))       # False
```

##### 7. Substring Check
- **in**: Its not a method but a keyword, it checks if a substring exists anywhere in the string and returns either true or false

```python
log = "2024-07-15 WARNING null value in column user_id"

print("WARNING" in log)     # True
print("ERROR" in log)       # False
print("user_id" in log)     # True
print("null value" in log)  # True - works on multi-word substrings

# Case sensitivity matters
print("warning" in log)             # False - case sensitive
print("warning" in log.lower())     # True - normalize first
```

##### 8. Locate a substring
- **.find() and .index()** both return the index(position) of the first occurence of a substring.
- Use **.find()** in data pipelines, since a missing substring should retuen -1

```python
s = "user_id=9921&region=APAC&active=True"

print(s.find("region"))         # 13 - position where "region" starts
print(s.find("xyz"))            # -1 - returns -1 if NOT found(safe)
print(s.index("region"))        # 13 same result as using .find()
print(s.index("xyz"))           # ValueError: substring not found (dangerous)
```

##### 9. Count occurences and zero-pad
- **.count() and .zfill()**

```python
log = "ERROR ERROR WARNING ERROR INFO"

print(log.count("ERROR"))   # 3

# zfill - zero-pad a numeric string to a fixed width
record_id = "42"

print(record_id.zfill(6))   # "000042" - useful for fixed-width file formats
```

##### 10. Chaining Methods
- Because each method retuns a new string you can chain them.
- The readability rule: Chain up to 3-4 methods max, if the chain gets longer, break it into named steps so its debuggable.

```python
raw = "  2024-07-15 | ERROR | DISK_FULL On node_7  "

result = raw.strip().lower().replace("_", " ")

print(result)      # 2024-07-15 | error | disk full on node 7
```

```python
# The readability rule.
# Hard to debug when something goes wrong.
raw = "  2024-07-15 | ERROR | DISK_FULL On node_7  "


result = raw.strip().split("|")[2].replace("_"," ").strip().title()

print(result)               # Disk Full On Node 7


print(raw)                  #   2024-07-15 | ERROR | DISK_FULL On node_7 

stripped = raw.strip()
print(stripped)             # 2024-07-15 | ERROR | DISK_FULL On node_7

parts = stripped.split("|")
print(parts)                # ['2024-07-15 ', ' ERROR ', ' DISK_FULL On node_7']

field = parts[2].strip()
print(field)                # DISK_FULL On node_7

normalized = field.replace("_", " ").title()
print(normalized)           # Disk Full On Node 7
```

```python
# CSV HEADER NORMALIZATION
raw_header = " User ID , First Name ,Last_Name,Date Of Birth , email Address ,Is Active  "

# Step 1: strip outer whitespaces from the full line
header = raw_header.strip()

print(raw_header)               #  User ID , First Name ,Last_Name,Date Of Birth , email Address ,Is Active  
print(header)                   # User ID , First Name ,Last_Name,Date Of Birth , email Address ,Is Active

# step 2: split into individual column names 
columns  = header.split(",")
print(columns)                  # ['User ID ', ' First Name ', 'Last_Name', 'Date Of Birth ', ' email Address ', 'Is Active']

# step 3: clean each column name
    # - strip whitespaces
    # - lowercase
    # replace spaces and hyphens with underscores
    
clean_columns = []
for col in columns:
    col = col.strip()               # remove surronding whitespace
    col = col.lower()               # normalize case
    col = col.replace(" ", "_")     # replace space with underscores
    col = col.replace("-", "_")     # replace hyphens with underscores
    clean_columns.append(col)       
    
print(clean_columns)                # ['user_id', 'first_name', 'last_name', 'date_of_birth', 'email_address', 'is_active']

# step 4: rebuild as a clean header string(for writing back to file)
clean_header = ",".join(clean_columns)

print(clean_header)                 #user_id,first_name,last_name,date_of_birth,email_address,is_active
```
